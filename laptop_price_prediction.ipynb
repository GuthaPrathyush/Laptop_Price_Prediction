```python
import re
def getProcessorName(Name):
    if re.search(r'i[357]', Name.split(' ')[-1]):
        return Name
    elif Name.split(' ')[0] == 'Intel':
        return 'Intel Other Processor'
    else:
        return 'AMD Processor'

df['Cpu_Name'] = df['Cpu_Name'].apply(getProcessorName)
```

```python
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
def changeTBtoGB(text):
    wordArray = text.split(' ')
    for i in range(len(wordArray)):
        if 'TB' in wordArray[i]:
            word = wordArray[i].strip()
            word = word[:-2]
            word = int(word)*1024
            wordArray[i] = f'{word}'
    return " ".join(wordArray)
df['Memory'] = df['Memory'].apply(changeTBtoGB)
df['Memory'] = df['Memory'].str.replace('GB', '')
sizeOfDF = len(df)
index = len(df.columns)-1
newValues = {
    'SSD': [0]*sizeOfDF,
    'HDD': [0]*sizeOfDF,
    'Flash_Storage': [0]*sizeOfDF,
    'Hybrid': [0]*sizeOfDF
}

for col_name, col_values in newValues.items():
    df.insert(index, col_name, col_values)
    index+=1

newDf = df['Memory'].str.split('+', n=1, expand=True)
newDf[1] = newDf[1].fillna('0')
newDf[0] = newDf[0].str.strip()
newDf[1] = newDf[1].str.strip()
first = newDf[0].str.findall(r'^([0-9]{0,4}.d*)').apply(lambda x: int(x[0]))
second = newDf[1].str.findall(r'^([0-9]{0,4}.d*)').apply(lambda x: int(x[0]))
for index, row in newDf.iterrows():
    if 'SSD' in row[0]:
        df.loc[index, 'SSD'] += first.loc[index]
    if 'HDD' in row[0]:
        df.loc[index, 'HDD'] += first.loc[index]
    if 'Flash Storage' in row[0]:
        df.loc[index, 'Flash_Storage'] += first.loc[index]
    if 'Hybrid' in row[0]:
        df.loc[index, 'Hybrid'] += first.loc[index]
    if 'SSD' in row[1]:
        df.loc[index, 'SSD'] += second.loc[index]
    if 'HDD' in row[1]:
        df.loc[index, 'HDD'] += second.loc[index]
    if 'Flash Storage' in row[1]:
        df.loc[index, 'Flash_Storage'] += second.loc[index]
    if 'Hybrid' in row[1]:
        df.loc[index, 'Hybrid'] += second.loc[index]
df.head(50)
```
